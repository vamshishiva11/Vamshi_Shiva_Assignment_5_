```python
# === 1. Install and Import Libraries ===
!pip install datasets xgboost wordcloud transformers scikit-learn --quiet
!pip install --upgrade transformers datasets --quiet

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import re
import nltk
import torch
import warnings

from wordcloud import WordCloud
from datasets import load_dataset, Dataset

from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer, CountVectorizer
from sklearn.preprocessing import LabelEncoder
from sklearn.naive_bayes import MultinomialNB
from sklearn.svm import LinearSVC
from sklearn.metrics import classification_report, confusion_matrix, accuracy_score, precision_score, recall_score, f1_score
from xgboost import XGBClassifier

from transformers import BertTokenizer, BertForSequenceClassification, Trainer, TrainingArguments
from transformers import DataCollatorWithPadding

nltk.download('stopwords')
nltk.download('wordnet')
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer

warnings.filterwarnings("ignore")

# === 2. Load and Balance Dataset ===
dataset = load_dataset("tweet_eval", "sentiment")
df = dataset["train"].to_pandas()
df = df.rename(columns={"text": "review", "label": "sentiment"})
df["sentiment"] = df["sentiment"].map({0: "negative", 1: "neutral", 2: "positive"})

# Balance the dataset
from sklearn.utils import resample
min_count = df["sentiment"].value_counts().min()
df_balanced = pd.concat([
    resample(df[df["sentiment"] == label], n_samples=min_count, replace=False, random_state=42)
    for label in df["sentiment"].unique()
])

# === 3. Preprocess Text ===
stop_words = set(stopwords.words("english"))
lemmatizer = WordNetLemmatizer()

def preprocess(text):
    text = re.sub(r"http\S+|@\w+|#\w+|[^a-zA-Z]", " ", text.lower())
    tokens = [lemmatizer.lemmatize(word) for word in text.split() if word not in stop_words]
    return " ".join(tokens)

df_balanced["cleaned"] = df_balanced["review"].apply(preprocess)
df_balanced["length"] = df_balanced["cleaned"].apply(lambda x: len(x.split()))

# === 4. EDA ===
sns.countplot(data=df_balanced, x="sentiment")
plt.title("Sentiment Class Distribution")
plt.show()

# WordClouds
for label in df_balanced["sentiment"].unique():
    text = " ".join(df_balanced[df_balanced["sentiment"] == label]["cleaned"])
    wordcloud = WordCloud(width=600, height=300).generate(text)
    plt.figure(figsize=(6, 3))
    plt.imshow(wordcloud, interpolation="bilinear")
    plt.axis("off")
    plt.title(f"{label.capitalize()} WordCloud")
    plt.show()

# Top N-grams
def plot_ngrams(corpus, ngram_range=(2,2), n=10, title='Top N-grams'):
    vec = CountVectorizer(ngram_range=ngram_range, stop_words="english").fit(corpus)
    bow = vec.transform(corpus)
    sum_words = bow.sum(axis=0)
    words_freq = [(word, sum_words[0, idx]) for word, idx in vec.vocabulary_.items()]
    words_freq = sorted(words_freq, key=lambda x: x[1], reverse=True)[:n]
    x, y = zip(*words_freq)
    sns.barplot(x=y, y=x)
    plt.title(title)
    plt.show()

plot_ngrams(df_balanced["cleaned"], (2,2), title="Top Bigrams")
plot_ngrams(df_balanced["cleaned"], (3,3), title="Top Trigrams")

# === 5. 80/20 Split & TF-IDF Vectorization ===
X = df_balanced["cleaned"]
y = df_balanced["sentiment"]
X_train, X_test, y_train, y_test = train_test_split(X, y, stratify=y, test_size=0.2, random_state=42)

vectorizer = TfidfVectorizer(max_features=1000)
X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)

label_encoder = LabelEncoder()
y_train_enc = label_encoder.fit_transform(y_train)
y_test_enc = label_encoder.transform(y_test)

# === 6. Traditional ML Models with 5-Fold CV + Evaluation ===
models = {
    "Naive Bayes": MultinomialNB(),
    "SVM (LinearSVC)": LinearSVC(C=1),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric="mlogloss")
}

summary = []

for name, model in models.items():
    print(f"\n=== Training Model: {name} ===")
    try:
        if name == "XGBoost":
            scores = cross_val_score(model, X_train_vec, y_train_enc, cv=5)
            model.fit(X_train_vec, y_train_enc)
            preds = model.predict(X_test_vec)
            preds = label_encoder.inverse_transform(preds)
        else:
            scores = cross_val_score(model, X_train_vec, y_train, cv=5)
            model.fit(X_train_vec, y_train)
            preds = model.predict(X_test_vec)

        print(f"\n{name} Classification Report:\n")
        print(classification_report(y_test, preds))
        sns.heatmap(confusion_matrix(y_test, preds), annot=True, fmt="d", cmap="Blues")
        plt.title(f"{name} Confusion Matrix")
        plt.xlabel("Predicted")
        plt.ylabel("Actual")
        plt.show()

        summary.append({
            "Model": name,
            "CV Accuracy": scores.mean(),
            "Test Accuracy": accuracy_score(y_test, preds),
            "Precision": precision_score(y_test, preds, average="weighted"),
            "Recall": recall_score(y_test, preds, average="weighted"),
            "F1 Score": f1_score(y_test, preds, average="weighted")
        })

    except Exception as e:
        print(f"❌ Failed model: {name} - Error: {str(e)}")

# === 7. Pretrained BERT (Fine-tuning) ===
bert_tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
df_bert = df_balanced[["review", "sentiment"]].rename(columns={"review": "text", "sentiment": "label"})
df_bert["label"] = df_bert["label"].map({"negative": 0, "neutral": 1, "positive": 2})

train_texts, test_texts = train_test_split(df_bert, stratify=df_bert["label"], test_size=0.2, random_state=42)
train_ds = Dataset.from_pandas(train_texts)
test_ds = Dataset.from_pandas(test_texts)

def tokenize(batch):
    return bert_tokenizer(batch["text"], padding=True, truncation=True)

train_ds = train_ds.map(tokenize, batched=True)
test_ds = test_ds.map(tokenize, batched=True)

data_collator = DataCollatorWithPadding(tokenizer=bert_tokenizer)
model = BertForSequenceClassification.from_pretrained("bert-base-uncased", num_labels=3)

training_args = TrainingArguments(
    output_dir="./results",
    per_device_train_batch_size=8,
    per_device_eval_batch_size=8,
    num_train_epochs=2,
    logging_dir="./logs",
    logging_steps=10
)

trainer = Trainer(
    model=model,
    args=training_args,
    train_dataset=train_ds,
    eval_dataset=test_ds,
    tokenizer=bert_tokenizer,
    data_collator=data_collator
)

print("\n=== Training BERT ===")
trainer.train()

preds = trainer.predict(test_ds)
bert_preds = np.argmax(preds.predictions, axis=1)

print("\nBERT Classification Report:\n")
print(classification_report(test_ds["label"], bert_preds))
sns.heatmap(confusion_matrix(test_ds["label"], bert_preds), annot=True, fmt="d", cmap="Purples")
plt.title("BERT Confusion Matrix")
plt.xlabel("Predicted")
plt.ylabel("Actual")
plt.show()

summary.append({
    "Model": "BERT",
    "CV Accuracy": np.nan,
    "Test Accuracy": accuracy_score(test_ds["label"], bert_preds),
    "Precision": precision_score(test_ds["label"], bert_preds, average="weighted"),
    "Recall": recall_score(test_ds["label"], bert_preds, average="weighted"),
    "F1 Score": f1_score(test_ds["label"], bert_preds, average="weighted")
})

# === 8. Final Model Comparison Table ===
summary_df = pd.DataFrame(summary)
print("\n=== Model Comparison Summary ===")
print(summary_df.to_string(index=False))
```

```python
# ✅ Final Sentiment Classifier with 80/20 Split + 10-Fold CV
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split, StratifiedKFold, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.pipeline import Pipeline
from sklearn.metrics import classification_report
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
import warnings
warnings.filterwarnings("ignore")

# === Load Data ===
def parse_data(filepath):
    data = []
    with open(filepath, "r") as f:
        for line in f:
            if line.strip():
                label, text = int(line[0]), line[2:].strip()
                data.append((label, text))
    return pd.DataFrame(data, columns=["label", "text"])

train_file = "/content/stsa-train.txt"
test_file = "/content/stsa-test.txt"

train_data = parse_data(train_file)
test_data = parse_data(test_file)

# === Initial EDA ===
print(f"Train shape: {train_data.shape}")
print(f"Test shape: {test_data.shape}")
print("Train label distribution:\n", train_data["label"].value_counts())
print("Test label distribution:\n", test_data["label"].value_counts())

# Plot label counts
sns.countplot(data=train_data, x="label")
plt.title("Train Label Distribution")
plt.show()

# Review length distribution
train_data["token_count"] = train_data["text"].apply(lambda x: len(x.split()))
sns.histplot(train_data["token_count"], bins=30)
plt.title("Review Word Count Distribution")
plt.show()

# === Train-Test Split ===
X_all = train_data["text"]
y_all = train_data["label"]

X_train, X_holdout, y_train, y_holdout = train_test_split(
    X_all, y_all, test_size=0.2, stratify=y_all, random_state=101
)

# === Classifiers and Pipeline Setup ===
vectorizer = TfidfVectorizer(max_features=5000, stop_words="english")
models = {
    "SVM": SVC(),
    "KNN": KNeighborsClassifier(),
    "Decision Tree": DecisionTreeClassifier(),
    "Random Forest": RandomForestClassifier(),
    "XGBoost": XGBClassifier(use_label_encoder=False, eval_metric="logloss")
}

results_summary = []
cv = StratifiedKFold(n_splits=10, shuffle=True, random_state=101)

# === Train and Evaluate Models ===
for model_name, model in models.items():
    print(f"\n🔍 Evaluating: {model_name}")
    pipeline = Pipeline([("tfidf", vectorizer), ("clf", model)])

    # Cross-validation on 80% train data
    cv_scores = cross_val_score(pipeline, X_train, y_train, cv=cv, scoring="accuracy")

    # Fit and predict on 20% holdout set
    pipeline.fit(X_train, y_train)
    y_preds = pipeline.predict(X_holdout)
    report = classification_report(y_holdout, y_preds, output_dict=True)

    results_summary.append({
        "Model": model_name,
        "CV Accuracy (Train 80%)": round(cv_scores.mean(), 4),
        "Validation Accuracy (20%)": round(report["accuracy"], 4),
        "Precision (Label=1)": round(report["1"]["precision"], 4),
        "Recall (Label=1)": round(report["1"]["recall"], 4),
        "F1 Score (Label=1)": round(report["1"]["f1-score"], 4)
    })

# === Results Comparison ===
results_df = pd.DataFrame(results_summary)
print("\n📊 Final Evaluation Summary:\n")
print(results_df)
```

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import PCA
from sklearn.cluster import MiniBatchKMeans, DBSCAN, AgglomerativeClustering
from sklearn.metrics import silhouette_score
from scipy.cluster.hierarchy import dendrogram, linkage

import nltk
from nltk.corpus import stopwords
from nltk.stem import WordNetLemmatizer
import re

import gensim
from gensim.models import Word2Vec

# Download necessary nltk data
nltk.download('stopwords')
nltk.download('wordnet')

# 📄 Load Dataset (skip bad lines if any)
df = pd.read_csv("/content/Amazon_Unlocked_Mobile.csv", on_bad_lines='skip', engine='python')

# Taking a sample of 10,000 rows for faster processing
df = df[['Reviews']].dropna().reset_index(drop=True).sample(n=10000, random_state=42)

# Preprocessing function
def preprocess(text):
    text = re.sub(r"[^a-zA-Z]", " ", text.lower())  # Remove non-alphabetic characters
    tokens = text.split()  # Tokenize by whitespace
    lemmatizer = WordNetLemmatizer()
    tokens = [lemmatizer.lemmatize(word) for word in tokens if word not in stopwords.words('english')]  # Lemmatize and remove stopwords
    return " ".join(tokens)

df['cleaned'] = df['Reviews'].apply(preprocess)
df = df[df['cleaned'].str.strip().astype(bool)]

# 📊 Plot Review Length Distribution
df['length'] = df['cleaned'].apply(lambda x: len(x.split()))
plt.figure(figsize=(10, 4))
sns.histplot(df['length'], bins=50)
plt.title('Review Length Distribution')
plt.xlabel('Number of Words')
plt.show()

# 🔎 TF-IDF Vectorization (using fewer features for speed)
tfidf = TfidfVectorizer(max_features=500)  # Reduced the number of features for faster processing
X_tfidf = tfidf.fit_transform(df['cleaned'])

# 🔑 Dimensionality Reduction with PCA
pca = PCA(n_components=2)
reduced = pca.fit_transform(X_tfidf.toarray())

# 🏷️ KMeans Clustering (Using MiniBatchKMeans for speed)
kmeans = MiniBatchKMeans(n_clusters=5, random_state=42, n_init=10, batch_size=1000, verbose=0)
kmeans_labels = kmeans.fit_predict(X_tfidf)

# 📊 Plot KMeans Clusters
plt.figure(figsize=(8,6))
plt.scatter(reduced[:, 0], reduced[:, 1], c=kmeans_labels, cmap='rainbow')
plt.title("KMeans Clusters")
plt.show()

# 🏷️ DBSCAN Clustering
dbscan = DBSCAN(eps=0.5, min_samples=5)
dbscan_labels = dbscan.fit_predict(X_tfidf)

# 📊 Plot DBSCAN Clusters
plt.figure(figsize=(8,6))
plt.scatter(reduced[:, 0], reduced[:, 1], c=dbscan_labels, cmap='rainbow')
plt.title("DBSCAN Clusters")
plt.show()

# 🏷️ Agglomerative Clustering
agg = AgglomerativeClustering(n_clusters=5)
agg_labels = agg.fit_predict(X_tfidf.toarray())

# 📊 Plot Agglomerative Clusters
plt.figure(figsize=(8,6))
plt.scatter(reduced[:, 0], reduced[:, 1], c=agg_labels, cmap='rainbow')
plt.title("Hierarchical Clusters")
plt.show()

# 📊 Dendrogram
linked = linkage(X_tfidf.toarray()[:100], method='ward')  # Limit the data to the first 100 rows for dendrogram
plt.figure(figsize=(10, 5))
dendrogram(linked, orientation='top', distance_sort='descending', show_leaf_counts=True)
plt.title("Word2Vec")
plt.show()

# Word2Vec model (training with sample size reduced for speed)
sentences = [text.split() for text in df['cleaned']]
model_w2v = Word2Vec(sentences, vector_size=100, window=5, min_count=2, workers=4)
w2v_features = []

# Vectorization of sentences
for sent in sentences:
    # If sentence has valid words for Word2Vec, calculate mean vector, else use zeros
    vector = np.mean([model_w2v.wv[word] for word in sent if word in model_w2v.wv], axis=0)
    if np.isnan(vector).any():  # If the sentence had no valid words, fill with zeros
        vector = np.zeros(100)
    w2v_features.append(vector)

# Convert to numpy array and handle NaNs
w2v_features = np.array(w2v_features)

# 🏷️ KMeans Clustering with Word2Vec features
w2v_kmeans = MiniBatchKMeans(n_clusters=5, random_state=42, n_init=10, batch_size=1000, verbose=0)
w2v_labels = w2v_kmeans.fit_predict(w2v_features)

# 📊 Plot Word2Vec KMeans Clusters
w2v_reduced = PCA(n_components=2).fit_transform(w2v_features)
plt.figure(figsize=(8,6))
plt.scatter(w2v_reduced[:, 0], w2v_reduced[:, 1], c=w2v_labels, cmap='rainbow')
plt.title("Word2Vec KMeans Clusters")
plt.show()
```
